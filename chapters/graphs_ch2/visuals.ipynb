```python
import numpy as np
import matplotlib.pyplot as plt

a = 10  # Arbitrary value for a
A = 2 * np.sqrt(3) / np.sqrt(a**3)

x1 = np.linspace(0, a/2, 100)  # Values of x in the first interval
x2 = np.linspace(a/2, a, 100)  # Values of x in the second interval

Psi1 = A * x1
Psi2 = A * (a - x2)

plt.plot(x1, Psi1, label='$\Psi$ in [0, a/2]')
plt.plot(x2, Psi2, label='$\Psi$ in [a/2, a]')

plt.xlabel('x')
plt.ylabel('$\Psi(x,0)$')
plt.title('Plot of $\Psi(x,0)$')

plt.legend()
plt.grid(True)
plt.savefig('2_7_1.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

# Constants
m = 1.0  # mass
omega = 1.0  # angular frequency
hbar = 1  # reduced Planck constant, but made legible with 1

# Define the wave function (psi2)
def psi2(x):
    coeff = np.sqrt(2.0 * m * omega / (np.pi * hbar))**0.25
    exp_term = np.exp(-m * omega / (2.0 * hbar) * x**2)
    return coeff * (m * omega * x**2 / hbar - 0.5) * exp_term

def psi1(x):
   coeff =  (m*omega/(np.pi * hbar))**0.25 * np.sqrt(2*m*omega/hbar)
   exp_term = np.exp(-m * omega / (2.0 * hbar) * x**2)
   return coeff * exp_term * x

def psi0(x):
    return (m*omega/(np.pi * hbar))**0.25 * np.exp(-m * omega / (2.0 * hbar) * x**2)
    
# Generate x values
x = np.linspace(-5, 5, 100)

# Calculate the wave function (psi2) for each x value
y_psi2 = psi2(x)

# Calculate the antiderivatives for each x value
y_psi1 = psi1(x)
y_psi0 = psi0(x)

# Create a single subplot with three plots side by side
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# Plot the wave function (psi2)
axes[0].plot(x, y_psi2)
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$\psi_2(x)$')

# Plot the first antiderivative (psi1)
axes[1].plot(x, y_psi1)
axes[1].set_xlabel('$x$')
axes[1].set_ylabel('$\psi_1(x)$')

# Plot the second antiderivative (psi0)
axes[2].plot(x, y_psi0)
axes[2].set_xlabel('$x$')
axes[2].set_ylabel('$\psi_0(x)$')

plt.suptitle('First 3 Wave Functions for Harmonic Oscillator')
plt.tight_layout()
plt.savefig('2_10_1.png')
plt.show()
```
